<b>Observations:</b>
<ol>
    <li>22% of the credit cards are going to be defaulted next month.</li>
    <li>The education column has a min value of 0, the meaning of which is not recorded.<br>
       Merge the category 0 into the "other" category, which is represented by 4.
    </li>
    <li>The marriage column has a min value of 0, the meaning of which is not recorded.<br>
       Merge the category 0 into the "other" category, which is represented by 3.
    </li>
    <li>The max value in the PAY columns is 8. There exists some customers who delayed their payments for 8 months.</li>
    <li>In the PAY columns, the mean is negative, indicating that most card holders pay their bills duly.<br>
       However, the minimum of the PAY columns is -2, the meaning of which is not recorded in the data dictionary.<br>
       Here, I assumed that -2 has the same meaning as -1, that the customer pays duly.<br>
       Moreover, the PAY columns can have a value of 0, the meaning of which is not recorded as well.
    </li>
    <li>The minimum of BILL_AMT columns are negative. There are cases in which the card issuer owes money to the card holder.</li>
</ol>

The education has 6 categories, but all of 4, 5, and 6 means "unknown." These categories can be merged into one.

```python
def map_education(category):
    if category >= 4:
        return 4
    else:
        return category

df.EDUCATION = df.EDUCATION.apply(map_education)
```

```python
def map_education_other(category):
    if category == 0:
        return 4
    else:
        return category

def map_marriage(category):
    if category == 0:
        return 3
    else:
        return category

df.EDUCATION = df.EDUCATION.apply(map_education_other)
df.MARRIAGE = df.MARRIAGE.apply(map_marriage)
```

<h2>Part2. Logistic Regression</h2>

<h3>2.1 Standardize Data</h3><br>
Since there are a lot of outliers in the numeric columns of the data, it is necessary to normalize the data before performing logistic regression.

```python
scaler = StandardScaler()
df_norm = pd.DataFrame(scaler.fit_transform(df))
df_norm.columns = df.columns
df_norm.head()
```

<h3>2.2 Learning Curve of the Logistic Regression Model</h3><br>

```python
num_size = 100
training_sizes = np.linspace(1/num_size,1,num_size)
```

```python
log_reg = LogisticRegression(random_state=42)
train_sizes, train_scores, validation_scores = learning_curve(
    estimator = log_reg,
    X = df_norm[cols], y = df['default'],
    train_sizes=training_sizes,
    shuffle=True, cv=5, 
    scoring = 'neg_mean_squared_error'
)

plt.subplots(figsize=(10,5))
plt.plot(train_sizes, -train_scores.mean(axis=1), label = 'Training error')
plt.plot(train_sizes, -validation_scores.mean(axis=1), label = 'Validation error')
plt.ylabel('Mean Squared Error')
plt.xlabel('Training Set Size')
plt.title('Learning curves for the Logistic Regression Model Using All Columns as Features')
plt.legend()
plt.show()
```

The error starts raising as the training set size increases. If all columns are used, the model is overfitting the data. The number of features used should be decreased.

<h3>2.3 Split Data</h3><br>
Use 80% of data as training and 20% as testing.

```python
X_norm_train, X_norm_test, y_norm_train, y_norm_test = train_test_split(df_norm[cols],
                                                                        df['default'],
                                                                        test_size=0.2, random_state=50)
```

<h3>2.4 Find Most Important Features Using RFE</h3><br>
Test different number of features to be selected using RFE, and find the number of features that produces the highest precision.

```python
log_reg = LogisticRegression(random_state=42)

max_feat = len(X_norm_train.columns)

precision = np.zeros(max_feat)

for i in range(0,max_feat):
    rfe = RFE(log_reg, n_features_to_select=i+1)
    rfe = rfe.fit(X_norm_train, y_norm_train)
    select_cols = list(np.array(X_norm_train.columns)[rfe.support_])
    df_select = df_norm[select_cols]
    X_sel_train, X_sel_test, y_sel_train, y_sel_test = train_test_split(df_select,
                                                                        df['default'],
                                                                        test_size=0.3, random_state=50)
    log_reg = LogisticRegression(random_state=50)
    log_reg.fit(X_sel_train,y_sel_train)
    precision[i] = log_reg.score(X_sel_test, y_sel_test)

n_feat = np.where(precision == np.amax(precision))[0][0]
```

```python
plt.subplots(figsize=(15,5))
plt.plot(range(1,max_feat+1),precision)
plt.ylabel('Precision')
plt.xlabel('Number of Columns Used as Features')
plt.title('Learning Curves for the Logistic Regression Model Using Different Number of Columns')
plt.show()
```

The precision of the model peaks when 1 column, 7 columns, or all columns are used as features.
